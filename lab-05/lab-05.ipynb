```python
item=["studs","football","cricket ball","bat","basketball","shuttle","grips","badminton shoes","chess board","strings"]
quantity=[6,4,7,9,4,5,9,4,9,3]
price=[235,567,345,678,453,63,890,356,444,425]

#print(item)
#print(quantity)
#print(price)
```

```python
import random # Importing the external library
i = 1
while i <= 100:
    r1 = random.randint(0, 9) # Generating the random number from the range and store it in r1 variable
    r2 = random.randint(0, 9) # Generating the random number from the range and store it in r2 variable
    r3 = random.randint(0, 9) # Generating the random number from the range and store it in r3 variable
    print("Random value for items",r1)
    print("Random value for quantity",r2)
    print("Random value for price",r3)

    it = item[r1]
    qt = quantity[r2]
    pr = price[r3]
    bill = qt*pr  #total bill 
    conc = it + '\t' + str(qt) + '\t' + str(pr) + '\t' + str(bill)+ '\n'    #adding the value in variable
    i=i+1       #iterating the value
    print(conc)     #printing the value 
    
    
    file = open("23122034.csv","+a")
    file.write(conc)
    file.close()
    

```
